```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('darkgrid')
%matplotlib inline
import numpy as np
```

```python
# Import data (Make sure to parse dates. Consider setting index column to 'date'.)
df = pd.read_csv('fcc-forum-pageviews.csv')
df.head()
```

## data cleaning

```python
# turning date column to datetime type
df['date'] = pd.to_datetime(df.date)
```

```python
# getting the top 2.5% views and bottom 2.5% views from the dateset
df = df[(df.value > np.percentile(df.value, 2.5)) & (df.value < np.percentile(df.value, 97.5))] 
```

```python
# creting months column 
df['month'] = df.date.dt.month_name()

# creting years column 
df['year'] = df.date.dt.year
df.head()
```

## data vis

### Visual 1 

```python
fig = plt.figure(figsize = (20,5))
plt.plot(df.date, df.value, color = 'red')
plt.title('Daily freeCodeCamp Forum Page Views 5/2016-12/2019')
plt.xlabel('Date')
plt.ylabel('Page Views');
```

### Visual 2

```python
fig = plt.figure(figsize = (10,7))
months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
main_color = sns.color_palette("Dark2", 12)
sns.barplot(x = df.year, y= df.value, hue=df.month, ci=None, hue_order = months_order, palette= main_color)
plt.xlabel('Years')
plt.ylabel('Average Page Views');
plt.legend(loc = 2, title="Months")
```

### Visual 3 

```python
fig, ax = plt.subplots(nrows =1 , ncols=2, figsize = (27,13))

months_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sns.boxplot(x = df.year, y=df.value, ax= ax[0] )
ax[0].set_ylabel('Page Views')
ax[0].set_xlabel('Year')
ax[0].set_title('Year-wise Box Plot (Trend)')

sns.boxplot(x = df.month, y=df.value, ax= ax[1], order= months_order )
ax[1].set_ylabel('Page Views')
ax[1].set_xlabel('Month')
ax[1].set_title('Month-wise Box Plot (Seasonality)')
ax[1].set_xticklabels([i[0:3] for i in months_order]);
```
